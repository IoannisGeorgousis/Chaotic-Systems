```python
from numpy import array,zeros
import numpy as np
import scipy as scipy
from scipy import integrate
import pylab as pl
```

```python
# Function for (a) set of constants
def F1(y,t,a=1.0,b=1.8):
    F1 = zeros(2)
    F1[0] = 1 - (1+b)*y[0] + a*y[0]**2*y[1]
    F1[1] = b*y[0]-a*y[0]**2*y[1]
    return F1

# Function for (b) set of constants
def F2(y,t,a=1.0,b=2.02):
    return F1(y, t, 1.0, 2.02)



y = array([0.0, 0.0]) # Initial conditions of x and y
t=np.linspace(0, 100, 10000) # Range of integration

z1=scipy.integrate.odeint(F1, y, t) # Solve for (a)
z2=scipy.integrate.odeint(F2, y, t) # Solve for (b)


# Plot x, y for(a)
f, axs = pl.subplots(2,2,figsize=(15,5))
pl.subplot(121)
pl.plot(t,z1[:,0], label='x')
pl.plot(t,z1[:,1], label='y') 
pl.xlabel('Time')
pl.ylabel(' ')
pl.title('X, y vs Time Plot with conditions (a)')
pl.legend()


# Plot x,y for (b)
pl.subplot(122)
pl.plot(t,z2[:,0], label='x')
pl.plot(t,z2[:,1], label='y') 
pl.xlabel('Time')
pl.ylabel(' ')
pl.title('X, y vs Time Plot with conditions (b)')
pl.legend()
pl.show()

print('\r' + 'From the difference in these two plots we see that for a very small change in b (of about 12.2%), the evolution ')
print('of the concentrations of catalysts changed drastically. In the first case, both x and y appear to converge to a certain ')
print('value (x = 1, y = 1.8 ) and mimic a damped oscillation, while after the change they continue to demonstrate oscillatory ')
print('behaviour about a value without ever converging.')
print('\r')


# Plot y vs x for (a)
f, axs = pl.subplots(2,2,figsize=(15,5))
pl.subplot(121)
pl.plot(z1[:,0], z1[:,1])
pl.xlabel('x')
pl.ylabel('y')
pl.title('Y vs x plot with conditions (a)')


#Plot y vs x for (b)
pl.subplot(122)
pl.plot(z2[:,0], z2[:,1])
pl.xlabel('x')
pl.ylabel('y')
pl.title('Y vs x plot with conditions (b)')
pl.show()

print('\r' + 'In the first plot we see how as x evolves with time, y tends closer and closer to a specific value (y=1.8) which  ')
print('is where x appears to converge on (at about x = 1)')
print('In the second plot we see that after the change of the b constant, y does not spiral into a certain value and simply ')
print('continues to oscillate in a manner around about y = 1.8, as x continues to remain around but not converge on ~1')
print('\r')
```
